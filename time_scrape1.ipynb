```python
from baseball_scraper import baseball_reference
from pybaseball import schedule_and_record
import pandas as pd
```

```python
from time import strptime

data = pd.DataFrame
data = schedule_and_record(2019, "CHC")
data['Date'] = data.apply(lambda data: str(data['Date']) + " " + str(2019), axis=1)
#data = pd.DataFrame()
for i in range(2020, 2025):
    #printf(i)
    curr = schedule_and_record(i, "CHC")
    curr['Date'] = curr.apply(lambda curr: str(curr['Date']) + " " + str(i), axis=1)
    data = pd.concat([data, curr])

data.head()
```
